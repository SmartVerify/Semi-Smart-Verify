## MULTIPLE IMAGES
converting multiple images to grey scale

```python
from PIL import Image
import os

def convert_to_grayscale(input_path, output_path):
    # Open the image file
    original_image = Image.open(input_path)

    # Convert the image to grayscale
    grayscale_image = original_image.convert('L')

    # Save the grayscale image with the correct file extension
    output_image_path = os.path.join(output_path, f"{os.path.splitext(os.path.basename(input_path))[0]}_greyscale.png")
    grayscale_image.save(output_image_path)

    print(f"Image converted to grayscale and saved at {output_image_path}")

def process_multiple_images(input_folder, output_folder):
    # Get a list of all image files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    # Process each image
    for image_file in image_files:
        input_image_path = os.path.join(input_folder, image_file)
        convert_to_grayscale(input_image_path, output_folder)

# Example usage
input_folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\data"
output_folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\greyscale"

process_multiple_images(input_folder_path, output_folder_path)
```

```python
from PIL import Image
import pytesseract
import os

def preprocess_image(image_path):
    """
    Open and preprocess the image file.

    Args:
    - image_path (str): The path to the image file.

    Returns:
    - Image: The preprocessed image.
    """
    image = Image.open(image_path)
    return image

def extract_text_from_image(image):
    """
    Use Tesseract to extract text from the image.

    Args:
    - image (Image): The preprocessed image.

    Returns:
    - str: Extracted text.
    """
    try:
        extracted_text = pytesseract.image_to_string(image)
        return extracted_text
    except Exception as e:
        print(f"Error extracting text: {e}")
        return ""

def extract_info_from_text(text, keywords):
    """
    Extract information from the text based on specified keywords.

    Args:
    - text (str): The extracted text.
    - keywords (list): List of keywords to search for.

    Returns:
    - dict: Extracted information.
    """
    extracted_info = {keyword: None for keyword in keywords}

    for keyword in keywords:
        keyword_position = text.find(keyword)
        if keyword_position != -1:
            start_index = keyword_position + len(keyword) + 1
            end_index = text.find('\n', start_index)
            value = text[start_index:end_index].strip()
            extracted_info[keyword] = value

    return extracted_info

def process_images_in_folder(folder_path):
    """
    Process all images in the specified folder and extract information.

    Args:
    - folder_path (str): The path to the folder containing images.
    """
    # List all files in the folder
    image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    # Define keywords
    keywords = ['Name', "Father's Name", 'Date of Birth', 'Signature', 'Permanent Account Number']

    for image_file in image_files:
        # Construct the full path to the image file
        image_path = os.path.join(folder_path, image_file)

        # Extract information from each image
        image = preprocess_image(image_path)
        extracted_text = extract_text_from_image(image)
        extracted_info = extract_info_from_text(extracted_text, keywords)

        # Print the results
        print(f"Extracted Information from {image_file}:")
        for key, value in extracted_info.items():
            print(f"{key}: {value}")
        print("=" * 30)

# Example usage
input_folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\data"
process_images_in_folder(input_folder_path)
```

```python
from PIL import Image, ImageEnhance, ImageFilter
import pytesseract
import os

def preprocess_image(image_path):
    try:
        with Image.open(image_path) as image:
            # Resize the image for better OCR accuracy
            image = image.resize((800, 600))
            
            # Convert the image to grayscale
            image = image.convert("L")
            
            # Apply contrast enhancement for better text visibility
            enhancer = ImageEnhance.Contrast(image)
            image = enhancer.enhance(2.0)

            # Apply image thresholding to convert to binary format
            image = image.point(lambda p: p > 128 and 255)

            # Apply Gaussian blur to reduce noise
            image = image.filter(ImageFilter.GaussianBlur(radius=1))

            return image.copy()
    except Exception as e:
        print(f"Error opening or preprocessing image {image_path}: {e}")
        return None

def extract_value(text, keyword):
    keyword_position = text.find(keyword)
    if keyword_position != -1:
        start_index = keyword_position + len(keyword) + 1
        end_index = text.find('\n', start_index)
        value = text[start_index:end_index].strip()
        return value
    return None

def extract_text_from_image(image_path):
    preprocessed_image = preprocess_image(image_path)

    if preprocessed_image:
        try:
            # Experiment with OCR configuration options
            extracted_text = pytesseract.image_to_string(preprocessed_image, config='--psm 6')

            # Define keywords
            keywords = ['Name', "Father's Name", 'Date of Birth', 'Signature', 'Permanent Account Number']

            # Initialize dictionary to store extracted information
            extracted_info = {keyword: extract_value(extracted_text, keyword) for keyword in keywords}

            return extracted_info
        except Exception as e:
            print(f"Error extracting text from image {image_path}: {e}")

    return None

def process_images_in_folder(folder_path):
    # List all files in the folder
    image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    for image_file in image_files:
        # Construct the full path to the image file
        image_path = os.path.join(folder_path, image_file)

        # Extract information from each image
        extracted_info = extract_text_from_image(image_path)

        if extracted_info:
            # Print the results
            print(f"Extracted Information from {image_file}:")
            for key, value in extracted_info.items():
                print(f"{key}: {value}")
            print("=" * 30)

# Example usage
input_folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\greyscale"
process_images_in_folder(input_folder_path)
```

```python
from PIL import Image, ImageEnhance
import pytesseract
import os
import logging

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

def preprocess_image(image_path):
    try:
        with Image.open(image_path) as image:
            # Resize the image for better OCR accuracy
            image = image.resize((800, 600))
            
            # Convert the image to grayscale
            image = image.convert("L")
            
            # Apply contrast enhancement for better text visibility
            enhancer = ImageEnhance.Contrast(image)
            image = enhancer.enhance(2.0)

            # Apply thresholding to convert to binary image
            threshold = 128
            image = image.point(lambda p: p > threshold and 255)

            return image.copy()
    except Exception as e:
        logger.error(f"Error opening or preprocessing image {image_path}: {e}")
        return None

def extract_value(text, keyword):
    keyword_position = text.find(keyword)
    if keyword_position != -1:
        start_index = keyword_position + len(keyword) + 1
        end_index = text.find('\n', start_index)
        value = text[start_index:end_index].strip()
        return value
    return None

def extract_text_from_image(image_path):
    preprocessed_image = preprocess_image(image_path)

    if preprocessed_image:
        try:
            extracted_text = pytesseract.image_to_string(preprocessed_image)

            # Define keywords
            keywords = ['Name', "Father's Name", 'Date of Birth', 'Signature', 'Permanent Account Number']

            # Initialize dictionary to store extracted information
            extracted_info = {keyword: extract_value(extracted_text, keyword) for keyword in keywords}

            return extracted_info
        except Exception as e:
            logger.error(f"Error extracting text from image {image_path}: {e}")

    return None

def process_images_in_folder(folder_path):
    # List all files in the folder
    image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    for image_file in image_files:
        # Construct the full path to the image file
        image_path = os.path.join(folder_path, image_file)

        # Extract information from each image
        extracted_info = extract_text_from_image(image_path)

        if extracted_info:
            # Print the results
            logger.info(f"Extracted Information from {image_file}:")
            for key, value in extracted_info.items():
                logger.info(f"{key}: {value}")
            logger.info("=" * 30)

# Example usage
input_folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\greyscale"
process_images_in_folder(input_folder_path)
```
