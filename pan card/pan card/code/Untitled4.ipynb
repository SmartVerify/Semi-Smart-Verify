```python
import os
import matplotlib.pyplot as plt

def disp_img(img, save_path=None, save_name=None):
    # Display the image
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    # Save the image if save_path and save_name are provided
    if save_path:
        # Check if save_path is a complete path or a folder
        if save_name:
            # If save_path is a complete path, use it as is
            save_path = os.path.dirname(save_path)
        save_name = f"{os.path.splitext(save_name)[0]}_converted.png"
        save_path = os.path.join(save_path, save_name)
        plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
        plt.close()
    else:
        plt.show()

def convert_and_display_images_from_folder(folder_path, save_folder_path=None):
    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"The specified folder '{folder_path}' does not exist.")
        return

    # Create a folder for saving images if save_folder_path is provided
    if save_folder_path and not os.path.exists(save_folder_path):
        os.makedirs(save_folder_path)

    # Get a list of all files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    if not image_files:
        print(f"No image files found in the folder '{folder_path}'.")
        return

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        # Read the image
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        if img is not None:
            # Display and save the image if save_folder_path is provided
            save_name = f"{os.path.splitext(image_file)[0]}_converted.png"
            save_path = os.path.join(save_folder_path, save_name) if save_folder_path else None
            disp_img(img, save_path=save_path, save_name=save_name)
        else:
            print(f"Error loading image: {image_path}")

# Example usage
folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\data"
save_folder_path = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\converted_images"
convert_and_display_images_from_folder(folder_path, save_folder_path)
```

```python
import os
import cv2
import matplotlib.pyplot as plt

def disp_img(img):
    plt.figure(figsize=(6, 4))
    plt.imshow(img, cmap='gray')  # Display grayscale image
    plt.axis('off')
    plt.show()

def apply_morphological_operations(folder_path, output_folder):
    # Check if the folder exists
    if not os.path.exists(folder_path):
        print(f"The specified folder '{folder_path}' does not exist.")
        return

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"Created output folder: {output_folder}")

    # Get a list of all files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    if not image_files:
        print(f"No image files found in the folder '{folder_path}'.")
        return

    # Define the kernel for morphological operations
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)

        # Read the image
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        if img is not None:
            # Apply morphological operations
            img_opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=1)

            # Save the processed image to the output folder
            output_path = os.path.join(output_folder, f"processed_{image_file}")
            cv2.imwrite(output_path, img_opening)
            
            # Display the result (optional)
            disp_img(img_opening)
        else:
            print(f"Error loading image: {image_path}")

# Example usage
input_folder = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\data"
output_folder = r"C:\Users\GP mysore\OneDrive\Desktop\pan card\processed_images"
apply_morphological_operations(input_folder, output_folder)
```

```python
import re

def extract_information_from_ocr(data):
    # Initialize variables
    name = fathers_name = dob = signature = pan = None

    # Iterate through the detected text
    i = 0
    while i < len(data['text']):
        text = data['text'][i].strip()

        # Extract 'Name'
        if 'Name' in text:
            name = re.search(r'Name\s*([\w\s]+)', text)
            name = name.group(1) if name else None

        # Extract "Father's Name"
        elif "Father's Name" in text:
            fathers_name = re.search(r"Father's Name\s*([\w\s]+)", text)
            fathers_name = fathers_name.group(1) if fathers_name else None

        # Extract 'DOB'
        elif re.match(r'\d{2}/\d{2}/\d{4}', text):
            dob = text

        # Extract 'Signature'
        elif 'Signature' in text:
            signature = 'present'

        # Extract 'PAN'
        elif re.match(r'[A-Z]{5}[0-9]{4}[A-Z]', text):
            pan = text

        i += 1

    return {'Name': name, 'Father\'s Name': fathers_name, 'DOB': dob, 'Signature': signature, 'PAN': pan}

# Example OCR text
ocr_text = "INCOME TAX DEPARTMENT GOVT OF INDIA carat Cal Aer HTS Permanent Account Number Card 4 KSXPK2621A aTa / Name KAKARLA SANDEEP frat I a1 / Father's Name RAMAKRISHNA KAKARLA ae ot rere J ae ' ate of Birth Soncep 30/08/2000 r FeATAT | Signature SS ay lage"

# Create a dummy OCR data dictionary
ocr_data = {'text': ocr_text.split('\n')}

# Extract information from OCR data
information = extract_information_from_ocr(ocr_data)

# Print the extracted information
print("Name:", information['Name'])
print("Father's Name:", information['Father\'s Name'])
print("DOB:", information['DOB'])
print("PAN:", information['PAN'])
print("Signature:", information['Signature'])
```

```python
import pandas as pd
import re

# Assuming the OCR_Result column contains the OCR results
# Replace this with the actual column name if it's different

# Define patterns for extraction
name_pattern = r'Name\s*:\s*([^"]+)'
fathers_name_pattern = r'Fathers\s*Name\s*:\s*([^"]+)'
dob_pattern = r'DOB\s*:\s*([^"]+)'
pan_pattern = r'PAN\s*:\s*([^"]+)'

# Function to extract information from a given row
def extract_information(row):
    name_match = re.search(name_pattern, row, flags=re.IGNORECASE)
    fathers_name_match = re.search(fathers_name_pattern, row, flags=re.IGNORECASE)
    dob_match = re.search(dob_pattern, row, flags=re.IGNORECASE)
    pan_match = re.search(pan_pattern, row, flags=re.IGNORECASE)

    name = name_match.group(1) if name_match else None
    fathers_name = fathers_name_match.group(1) if fathers_name_match else None
    dob = dob_match.group(1) if dob_match else None
    pan = pan_match.group(1) if pan_match else None
    
    return pd.Series({'Name': name, 'Fathers_Name': fathers_name, 'DOB': dob, 'PAN': pan})

# Apply the function to extract information for each row
df_extracted = df['Detected Text'].apply(extract_information)

# Concatenate the extracted information with the original DataFrame
df_combined = pd.concat([df, df_extracted], axis=1)

# Display the resulting DataFrame
print(df_combined)
```
